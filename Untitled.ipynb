```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
import os
import cv2

from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Model
from keras.layers import Input, Conv2D, BatchNormalization, Activation, UpSampling2D, Concatenate, Flatten, Dense, Reshape, MaxPooling2D, Conv2DTranspose
from tensorflow.keras.callbacks import EarlyStopping

%matplotlib inline
```

```python
import os
```

```python
train_image_path="../Denoising Documents AutoEncoders/denoising-dirty-documents/train/train/"
train_cleaned_path="../Denoising Documents AutoEncoders/denoising-dirty-documents/train_cleaned/train_cleaned"
test_path="../Denoising Documents AutoEncoders/denoising-dirty-documents/test/test"
train_imgs_name = sorted(os.listdir(train_image_path))
train_cleaned_imgs_name = sorted(os.listdir(train_cleaned_path))
test_imgs_name = sorted(os.listdir(test_path))
```

```python
def preprocessing(image_path):
    img=cv2.imread(train_image_path+'101.png')
    img = np.asarray(img, dtype="float32")
    img = cv2.resize(img, (480, 480))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = img/255.0
    img = np.reshape(img, (480, 480, 1))
    return img
```

```python
train_images=[os.path.join(train_image_path, img_name) 
             for img_name in train_imgs_name]
test_images=[os.path.join(test_path, img_name) 
            for img_name in test_imgs_name]
train_cleaned=[os.path.join(train_cleaned_path, img_name) 
             for img_name in train_cleaned_imgs_name]
```

```python
print("Len of train images: ", type(train_imgs_name[0]))
```

```python
img= cv2.imread(train_images[1])
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
axes[0].imshow(img)
axes[1].imshow(cv2.imread(train_cleaned[1]))
plt.show()
```

```python
train_img=[]
train_clean_img=[]
test_img=[]

for img_path in train_images:
    train_img.append(preprocessing(img_path))
for img_path in train_cleaned:
    train_clean_img.append(preprocessing(img_path))
for img_path in test_images:
    test_img.append(preprocessing(img_path))
```

```python
x_train = np.asarray(train_img)
y_train=np.asarray(train_clean_img)
x_test = np.asarray(test_img)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.15)
```

```python
print("Len of train set = ", len(x_train))
```

```python
class AutoencoderMultilayersWithPooling(Model):
    def __init__(self, encoding_dim):
        super(AutoencoderMultilayersWithPooling, self).__init__()
        
        total_layers = 40
        n_layers = int(total_layers/2)

        """
          Input of encoder: (480x480x1)
        """
        # encoder
        self.encoder_layers = []
        k = 32
        for i in range(n_layers):
            self.encoder_layers.append(Conv2D(k, (3,3), padding='same'))
            self.encoder_layers.append(BatchNormalization())
            self.encoder_layers.append(Activation('relu'))
            if i % 2 == 0 and i < 6:
                self.encoder_layers.append(MaxPooling2D((2, 2), padding='same')) 
                k *= 2

        """
          Input of bottlenect: (4x4x128)
        """
        # Bottleneck layer
        self.bottleneck_layers = [
            Flatten(),
            Dense(encoding_dim, activation='relu')
        ]

        """
          Input of decoder: eg: (encoding_dim,)
        """
        # Decoder Layers
        # decoder
        self.decoder_layers = [
            Dense(30 * 30 * 32, activation='relu'),
            Reshape((30, 30, 32)),
            Conv2DTranspose(32, (3,3), strides=(2, 2), padding='same')
        ]
        for i in range(n_layers):
            self.decoder_layers.append(Conv2D(32, (3,3), padding='same'))
            self.decoder_layers.append(BatchNormalization())
            self.decoder_layers.append(Activation('relu'))

        self.decoder_layers.append(Conv2D(1, (3,3), padding='same', activation='sigmoid'))

        #
        self.ConvTrans = []
        self.concat_layers = []
        for i in range(int(k/2) + 1):
            self.concat_layers.append(Concatenate())
            self.ConvTrans.append(Conv2DTranspose(32, (3,3), strides=(2, 2), padding='same'))

    def call(self, inputs):
        x = inputs
        skip_connections = []
        k = 0

        for i, layer in enumerate(self.encoder_layers):
            x = layer(x)
            if isinstance(layer, Activation):
                if k % 2 == 0:
                    skip_connections.append(x)
                k += 1

        for layer in self.bottleneck_layers:
            x = layer(x)

        N_decoder_layer = len(self.decoder_layers)
        l = 0
        p = 0
        q = 0
        for i, layer in enumerate(self.decoder_layers):
            if (i == N_decoder_layer-4) or (i == N_decoder_layer-10) or (i == N_decoder_layer-16):
                x = self.ConvTrans[q](x)
                q += 1
#                 x = self.concat_layers[p]([x, skip_connections.pop(-1)]) #  just for 5 layers
#                 p += 1
            # just for 10, 20, 30 layers
            if isinstance(self.decoder_layers[i-1], Activation):
                if l % 2 == 0:
#                 if l % 2 == 1: # just for 5 layers
                    x = self.concat_layers[p]([x, skip_connections.pop(-1)]) # skip-connection
                    p += 1
            x = layer(x)
            if isinstance(layer, Activation):
                l += 1
        return x

    def create_model(self):
        x = Input(shape=(480, 480, 1))
        model = Model(inputs=[x], outputs=self.call(x))
        return model
```

```python
test_model = AutoencoderMultilayersWithPooling(encoding_dim=512).create_model()
test_model.build(input_shape=(None, 480, 480, 1))
```

```python
train_model = AutoencoderMultilayersWithPooling(encoding_dim=512)
train_model.compile(optimizer='adam', loss="MeanSquaredError")
```

```python
history = train_model.fit(x_train, y_train,
              batch_size=1,
              epochs=3,
              shuffle=True,
              validation_data=(x_val, y_val))
```

```python
y_test = train_model.predict(x_test, batch_size=8)
```
